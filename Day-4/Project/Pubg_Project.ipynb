```python
#How many unique values are there in 'matchType' and what are their counts?
cnt = 0
visited = []
for i in range(0, len(pubg['matchType'])): 
    
    if pubg['matchType'][i] not in visited:  
        
        visited.append(pubg['matchType'][i]) 
          
        cnt += 1
  
print("No.of.unique values :",cnt) 
  
print("unique values :",visited)
```
